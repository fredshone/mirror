```python
import pandas as pd
```

```python
census_sample = pd.read_csv("data/census.csv.zip")
census_sample = census_sample.set_index("resident_id_m")
census_regions = set(census_sample.region)
census_sample.head()
```

```python
nomis_projection = pd.read_csv("data/nomis/2026.tsv.zip", sep="\t")[
    ["geogcode", "Gender", "Age", "value"]
]
projection_zones = set(nomis_projection.geogcode)
nomis_projection.head()
```

```python
region_lookup = pd.read_csv("data/2019_region_lookup.csv")
# https://geoportal.statistics.gov.uk/datasets/e75d51eca981493992103772f6004e45_0/about
region_map = dict(zip(region_lookup.LAD19CD, region_lookup.RGN19CD))
lookup_zones = set(region_lookup.LAD19CD)
lookup_regions = set(region_lookup.RGN19CD)
assert (
    projection_zones == lookup_zones
), "Data zones and lookup zones do not match"
# assert (
#     census_regions == lookup_regions
# ), "Census regions and lookup regions do not match"
region_lookup.head()
```

```python
nomis_projection["region"] = nomis_projection.geogcode.map(region_map)
print(nomis_projection.region.isna().sum(), "missing age groups")
nomis_projection.head()
```

```python
nomis_projection["age_group"] = nomis_projection["Age"].map(
    {
        "Age 0 - 4": "0-15",
        "Aged 5-9": "0-15",
        "Aged 10-14": "0-15",
        "Aged 15-19": "16-24",
        "Aged 20-24": "16-24",
        "Aged 25-29": "25-34",
        "Aged 30-34": "25-34",
        "Aged 35-39": "35-44",
        "Aged 40-44": "35-44",
        "Aged 45-49": "45-54",
        "Aged 50-54": "45-54",
        "Aged 55-59": "55-64",
        "Aged 60-64": "55-64",
        "Aged 65-69": "65+",
        "Aged 70-74": "65+",
        "Aged 75-79": "65+",
        "Aged 80-84": "65+",
        "Aged 85+": "65+",
    }
)

print(nomis_projection.age_group.isna().sum(), "missing age groups")

nomis_projection.head()
```

```python
nomis_projection = nomis_projection.rename(
    columns={"geogcode": "zone", "Age": "age", "Gender": "sex"}
)
nomis_projection = nomis_projection.drop(columns=["age"])
nomis_projection.sex = nomis_projection.sex.map({"Male": 2, "Female": 1})
nomis_projection.head()
```
